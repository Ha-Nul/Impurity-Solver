```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import cm
from matplotlib.gridspec import GridSpec
from matplotlib.patches import Rectangle
```

```python
gs = GridSpec(nrows=1, ncols=2, width_ratios=[1,1], height_ratios=[1])
cmaps = plt.colormaps()
blues = cm.get_cmap("YlOrBr",lut=15)
spec = cm.get_cmap("Spectral",lut =15)
c_blues = [blues(i) for i in range(17)]
c_spec = [spec(i) for i in range(17)]
```

```python
beta = [3.3,3.57,3.84,4.16,4.54,5,5.55,6.25,7.14,8.33,10,12.5,16.66,25]
```

```python
gamar = np.linspace(0,1,21)
garr = []

for i in range(len(gamar)):
    if i == 0 or i == 20:
        garr.append(int(gamar[i]))
    else:
        garr.append(round(gamar[i],2))


aarr = [0,5,20]

alp_index = 1
```

```python
for i in beta:
    for j in aarr:
        globals()['data{}_a{}'.format(i,j)] = pd.read_csv('./TOA_beta{}_alp_{}.txt'.format(i,j),header=None,delim_whitespace=True)
```

```python
fig = plt.figure(figsize=(9,7))
mpl.rcParams['axes.linewidth'] = 3

plt.xlabel(r'$\gamma$')

plt.tick_params(axis='x',direction='in')
plt.tick_params(axis='y',direction='in')

plt.rcParams["font.family"] = "Helvetica"
plt.rcParams.update({'font.size': 11})

#plt.title(r"TOA result in $\alpha$ = " + str(aarr[alp_index]))

plt.xlabel(r'$\gamma$',fontsize=20)
plt.ylabel(r'$\beta\chi_{sp}(\tau=\frac{\beta}{2})$',fontsize=20)

#plt.xlabel(r"$\gamma$",style = "italic",fontsize=20)
#plt.ylabel("$\it{T}$",fontsize=20)
plt.text(0.02, 0.295, r'TOA',fontsize=25)
#plt.ylabel(r"Temp [1/$\beta$]")
#plt.xlim([0,1])
#plt.ylim([0.0,0.35])

plt.tick_params(axis='x',direction='in',size=10)
plt.tick_params(axis='y',direction='in',size=10)
plt.xticks(fontsize=18) 
plt.yticks(fontsize=18) 

#plt.grid()
plt.legend(fontsize=25,loc='lower right')

a = 0
for i in beta:
    plt.plot(garr,globals()['data{}_a{}'.format(i,aarr[alp_index])][1],label=r'$\beta = $' + str(i),color=c_blues[a+2],marker='x',markersize = 15 , linewidth = 2)
    a+=1

#plt.grid()
plt.legend(ncol=2,fontsize=18)
#plt.show()
plt.savefig("/Users/e2_602_qma/crosstest.pdf",format='pdf',bbox_inches='tight')
```
